```python
from curses import raw


def A_7():
    tau = 50*10**-6
    T = 1 / 1000
    #               n, nu_n^exp, nu_n^theor, a_n^exp, a_n^theor
    raw_data = [[i[0], [i[1], 0.01], [i[2]], [i[3], 1]] for i in pd.read_csv("data/A.7.csv").to_numpy()]
    a_n_theor = [[abs(math.sin(np.pi*i[0]*tau/T))/(np.pi*i[0])] for i in raw_data]


    for i in range(len(raw_data)):
        raw_data[i].append(countWithSigma(raw_data[i][3][0]/raw_data[0][3][0], [raw_data[i][3], raw_data[0][3]]))
        raw_data[i].append([a_n_theor[i][0]/a_n_theor[0][0]])

    headers = [r"$n$", r"$\nu_n^{эксп}, кГц$", r"$\nu_n^{теор}, кГц$", r"$\vert a_n \vert^{эксп}, усл. ед.$", r"$\vert a_n/a_1 \vert^{эксп}$", r"$\vert a_n/a_1 \vert^{теор}$"]

    table = []

    for i, h in enumerate(headers):
        if i == 0:
            table.append([h] + [int(j[i]) for j in raw_data[:5]])
        else:
            table.append([h] + [texAns(j[i]) for j in raw_data[:5]])

    texTable(table)

    table2 = []

    for i, h in enumerate(headers):
        if i == 0:
            table2.append([h] + [int(j[i]) for j in raw_data[5:]])
        else:
            table2.append([h] + [texAns(j[i]) for j in raw_data[5:]])

    texTable(table2)
A_7()
```
